# Replace

```python
df0["sex"]=df0["sex"].replace({"male":0,"female":1})
df0["region"]=df0["region"].replace(["southwest","southeast","northeast","northwest"],[0,1,2,3])
```

```python
df0["smoker"]=df0["smoker"].replace({"yes":0,"no":1})
```

```python
df1.sex=lab_en.fit_transform(df1.sex)
```

```python
df1.region=lab_en.fit_transform(df1.region)
df1.smoker=lab_en.fit_transform(df1.smoker)
```

# OneHot Encoding

## [Reference](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html#)

```python
df3.head()
```

```python
cate_data=["sex","smoker","region"]
onehot = OneHotEncoder(sparse_output=False,drop="first")
encode_data=onehot.fit_transform(df3[cate_data])
encode_column=onehot.get_feature_names_out(cate_data)
new_df=pd.DataFrame(encode_data,columns=encode_column)
new_df.head()
```

```python
drop_column=df3.drop(columns=cate_data)
df_onehot=pd.concat([drop_column,new_df],axis=1)
df_onehot.head()
```

# OnehotEncoding Using For Loop

```python
df8.head()
```

```python
for m in cate_data:
  col_encode=onehot.fit_transform(df8[[m]])
  new_df=pd.DataFrame(col_encode,columns=onehot.get_feature_names_out([m]))
  df8 = pd.concat([df5, new_df], axis=1)
  df8.drop(m, axis=1, inplace=True)


df8.head() 
```

# Get Dummies Using For Loop

```python
df5.head()
```

```python
my_catego_data=["sex","smoker","region"]
for n in my_catego_data:
  dumi=pd.get_dummies(df5[n],prefix=n,drop_first=True)
  df5=pd.concat([df5,dumi],axis=1)
  df5.drop(n, axis=1, inplace=True)
df5.head()
```
